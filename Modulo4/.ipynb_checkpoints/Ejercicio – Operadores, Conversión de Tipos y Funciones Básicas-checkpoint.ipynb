```python
lista = []
tupla = []
dic = {}
resultado = 0
#--------------------------------------------------------------------
def ingresarNum( lista, tupla, dic):
    for x in range(15):
        num = input('{}).- Ingrese un número: '.format(x+1))
        num = int(num)
        if x < 5:
            lista.append(num)
        elif 5 <= x < 10:
            tupla.append(num)
        else:
            clave = str(x)
            valor = num
            dic[clave] = valor
    return tuple(tupla)
#--------------------------------------------------------------------
def mostrar(resultado):
    print('El resultado de la operación es: {}'.format(resultado))
    print('El resultado de la operación en binario es: {}'.format(bin(resultado)))
    print('El resultado de la operación en octal es: {}'.format(oct(resultado)))
    print('El resultado de la operación en hexadecimal es: {}'.format(hex(resultado)))        
#--------------------------------------------------------------------
def suma(datos):
    resultado = 0
    if type(datos) == list:
        resultado=sum(datos)
        return resultado
    elif type(datos) == tuple:
        for x in datos:
            resultado+=x
        return resultado
    else:
        for x in datos.values():
            resultado+=x
        return resultado
#--------------------------------------------------------------------
#--------------------------------------------------------------------
def resta(datos):
    resultado = 0
    if type(datos) == list:
        for x in datos:
            resultado-=x
        return resultado
    elif type(datos) == tuple:
        for x in datos:
            resultado-=x
        return resultado
    else:
        for x in datos.values():
            resultado-=x
        return resultado
#--------------------------------------------------------------------
def producto(datos):
    resultado = 1
    if type(datos) == list:
        for x in datos:
            resultado*=x
        return resultado
    elif type(datos) == tuple:
        for x in datos:
            resultado*=x
        return resultado
    else:
        for x in datos.values():
            resultado*=x
        return resultado
#--------------------------------------------------------------------
#--------------------------------------------------------------------
def division(datos):
    if type(datos) == list:
        resultado=datos[0]
        for x in datos[1:]:
            resultado/=x
        return int(resultado)
    elif type(datos) == tuple:
        resultado=datos[0]
        for x in datos[1:]:
            resultado/=x
        return int(resultado)
    else:
        resultado = datos['10']
        for clave,valor in datos.items():
            if clave!='10':
                resultado/=valor
        return int(resultado)
#--------------------------------------------------------------------
tupla = ingresarNum(lista, tupla, dic)
tupla = tuple(tupla)
#--------------------------------------------------------------------
operacion = int(input('\nIngrese el número de la operación que desea hacer: SUMA (1) - RESTA (2) - PRODUCTO (3) - DIVISIÓN (4)'))
if(operacion == 1):
    print('Operación a realizar: SUMA\n\n')
    print('Valores ingresados en la LISTA: {}\n'.format(lista))
    resultado=suma(lista)
    mostrar(resultado)
    print('\nValores ingresados en la TUPLA: {}\n'.format(tupla))
    resultado=suma(tupla)
    mostrar(resultado)
    print('\nValores ingresados en el DICCIONARIO: {}\n'.format(dic))
    resultado=suma(dic)
    mostrar(resultado)
elif(operacion == 2):
    print('Operación a realizar: RESTA\n\n')
    print('Valores ingresados en la LISTA: {}\n'.format(lista))
    resultado=resta(lista)
    mostrar(resultado)
    print('\nValores ingresados en la TUPLA: {}\n'.format(tupla))
    resultado=resta(tupla)
    mostrar(resultado)
    print('\nValores ingresados en el DICCIONARIO: {}\n'.format(dic))
    resultado=resta(dic)
    mostrar(resultado)
elif(operacion == 3):
    print('Operación a realizar: PRODUCTO\n\n')
    print('Valores ingresados en la LISTA: {}\n'.format(lista))
    resultado=producto(lista)
    mostrar(resultado)
    print('\nValores ingresados en la TUPLA: {}\n'.format(tupla))
    resultado=producto(tupla)
    mostrar(resultado)
    print('\nValores ingresados en el DICCIONARIO: {}\n'.format(dic))
    resultado=producto(dic)
    mostrar(resultado)
elif(operacion == 4):
    print('Operación a realizar: DIVISION\n\n')
    print('Valores ingresados en la LISTA: {}\n'.format(lista))
    resultado=division(lista)
    mostrar(resultado)
    print('\nValores ingresados en la TUPLA: {}\n'.format(tupla))
    resultado=division(tupla)
    mostrar(resultado)
    print('\nValores ingresados en el DICCIONARIO: {}\n'.format(dic))
    resultado=division(dic)
    mostrar(resultado)
else:
    print('Operación no valida')
```
